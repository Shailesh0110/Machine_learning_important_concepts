```python
df.drop(columns=['PassengerId','Sex', 'Age', 'Ticket', 'Fare', 'Cabin', 'Embarked'],inplace=True)
```

```python
df['Family_size']=df['SibSp']+df['Parch']+1
```

```python
df
```

```python
def my_fun(num):
    if num==1:
        return 0 #alone
    if num>1 and num<=4:
        return 1 #small family
    else:
        return 2 # large family
```

```python
df['family_type']=df['Family_size'].apply(my_fun)
```

```python
df
```

```python
df.drop(columns=['Family_size','Parch', 'SibSp', 'Name'],inplace=True)
```
