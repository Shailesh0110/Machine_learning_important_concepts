```python
df=pd.read_csv("Titanic-Dataset.csv")
```

```python
df['Isfemale']=df['Sex'].map({'male':0,'female':1})
```

```python
df[['embC','embQ','embS']]=pd.get_dummies(df['Embarked'])
```

```python
X=df.drop(columns=['Survived','Name','Sex','Ticket','Cabin','Embarked'])
y=df['Survived']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25)
```

```python
X_test
```

```python
imputer=SimpleImputer(
    strategy='mean')
```

```python
scaler=StandardScaler()
```

```python
X_train, X_test, y_train, y_test 
```

```python
X_train['Age']=imputer.fit_transform(X_train[['Age']])
X_train[['Age','Fare']]=scaler.fit_transform(X_train[['Age','Fare']])
```

```python
X_test
```

```python
X_test['Age']=imputer.transform(X_test[['Age']])
X_test[['Age','Fare']]=scaler.transform(X_test[['Age','Fare']])
```

```python
model1=LogisticRegression()
model1.fit(X_train,y_train)
```

```python
y_pred1=model1.predict(X_test)
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred1)
```

```python
cv=RepeatedStratifiedKFold( n_splits=10, n_repeats=5)
cross_val_score(model1,X_train,y_train,cv=cv,scoring='accuracy').mean()
```

```python
print(confusion_matrix(y_test,y_pred1))
print(accuracy_score(y_test,y_pred1))
print(classification_report(y_test,y_pred1))
```

```python
model=KNeighborsClassifier(n_neighbors=3,weights="uniform")
model.fit(X_train,y_train)
```

```python
y_pred=model.predict(X_test)
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```

```python
cross_val_score(model,X_train,y_train,cv=5,scoring='accuracy').mean()
```
